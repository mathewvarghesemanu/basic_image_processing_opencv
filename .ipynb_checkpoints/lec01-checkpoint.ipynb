```python
# Load image
import cv2
import matplotlib
import matplotlib.pyplot as plt

im = cv2.imread('picket_fence.jpg')
h, w = im.shape[:2]

print(type(im))

print(im.shape, '   # Image is in HWC format')
print(im[0,0], '   # Each pixel is in BGR format')

plt.figure()
plt.imshow(im[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
```

```python
# dir
print(dir(im))
```

```python
# Examine memory -- https://numpy.org/doc/stable/reference/arrays.ndarray.html
print(type(im), '   # Image is an n-dimensional array')
print(im.dtype, '   # Each pixel is a byte')
print(type(im.shape), '   # Image shape is a tuple')
print(im.shape, '   # Dimensions are HWC')
print(im.strides, '   # Stride for H, W, C')
print(im.size, '   # Number of elements in array')
print(im.itemsize, '   # Number of bytes per element')
print(im.nbytes, '   # Total number of bytes in array')
print(type(im.data))
print(im.flags)  # C-style (row-major), not F-style (column-major)
```

```python
# BGR values of first 4 pixels, accessed as a 1-D array (Note: storage is row-major order.)
im2_rgb = cv2.imread('colors.png')
print(im2_rgb.shape, '   # HWC of small color image')
print(im2_rgb.flat[0:12], '   # Pixel colors are stored interleaved, in row-major order')

f = im2_rgb.flatten()
print(f[0:12], '   # First 12 pixels again')
print(f.shape, '   # Number of pixel values = H * W * C')

plt.imshow(im2_rgb[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
```

```python
# Slicing notation - https://stackoverflow.com/questions/509211/understanding-slice-notation

# These are all equivalent -- nothing changes
im2 = im2_rgb[:, :, :]  
im2 = im2_rgb[:, :, [0,1,2]]  
im2 = im2_rgb[:, :, 0:3]  # color channels 0, 1, 2 (not including 3)
im2 = im2_rgb[:, :, 0:3:1]  # step 1
im2 = im2_rgb[:, :, 0:99]  # automatically stops when reaches end of array
im2 = im2_rgb[:, :, 0:99:1]  
im2 = im2_rgb[:, :, -3:3]  # 3rd from the right is same as 0
im2 = im2_rgb[:, :, -3:3:1] 
im2 = im2_rgb[:, :, -3:99]
im2 = im2_rgb[:, :, -3:99:1]

plt.imshow(im2[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
```

```python
# Fix color by reversing color channels

# These are all equivalent
im2 = im2_rgb[:, :, ::-1]  # step -1
im2 = im2_rgb[:, :, [2,1,0]]  # color channels 2, 1, 0
im2 = im2_rgb[:, :, 2::-1]
im2 = im2_rgb[:, :, 2:-4:-1]
im2 = im2_rgb[:, :, 2:-99:-1]
im2 = im2_rgb[:, :, 99::-1]
im2 = im2_rgb[:, :, 99:-99:-1]
im2 = im2_rgb[:, :, -1:-4:-1]
im2 = im2_rgb[:, :, -1:-99:-1]

plt.imshow(im2[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
```

```python
# Reshape preserves the order of the pixel values
im2a = im2.reshape(12,4,3)
print(im2a.shape, '   # Shape of resized image')
plt.imshow(im2a[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
im2b = im2.reshape(12,12)
print(im2b.shape, '   # Shape of resized image')
plt.figure()
plt.imshow(im2b, cmap='gray')

# Equivalently
im2b = im2.reshape(12,12,1)
print(im2b.shape, '   # Shape of resized image -- before squeezing (imshow will throw error)')
# plt.imshow(im2b, cmap='gray') #  throws error
im2b = im2b.squeeze()  # Squeezing removes dimensions of 1
print(im2b.shape, '   # Shape of resized image -- after squeezing')
```

```python
# Views - Be careful when manipulating data in Python

im2c = im2.copy()  # Make a copy of the data
im2d = im2c
print(im2[0,0], '   # Originally, all values are the same')
print(im2c[0,0])
print(im2d[0,0], '\n')

im2c[0,0] = (6,7,8)
print(im2[0,0], '   # im2c and im2d are views to the same underlying data')
print(im2c[0,0])
print(im2d[0,0], '\n')

im2d[0,0] = (22,88,127)
print(im2[0,0], '   # im2c and im2d are views to the same underlying data')
print(im2c[0,0])
print(im2d[0,0])
```

```python
im2c = im2.copy()  # Make a copy of the data
im2d = im2c.ravel()
print(im2d.shape)
im2e = im2c.flatten()
print(im2e.shape, '\n')

print(im2[0,0], '   # `before changing values')
print(im2c[0,0])
print(im2d[0:3])
print(im2e[0:3], '\n')

im2d[0] = 57
print(im2[0,0], '   # `ravel` provides a view; `flatten` copies the data')
print(im2c[0,0])
print(im2d[0:3])
print(im2e[0:3], '\n')
```

```python
# Transpose, axes, etc.
im2f = im2.transpose()
plt.imshow(im2[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
print(im2.shape, im2f.shape, '   # Weird things happen')

im2f = im2.transpose((1,0,2))
plt.figure()
plt.imshow(im2f[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
print(im2.shape, im2f.shape, '   # Swapped row and column axes')

print(im2_rgb.strides, im2.strides, im2f.strides, '   # Strides')

im2s = im2[:2,:4,:]
print('Small image:')
print(im2s.shape)
print(im2s)
a = im2s.min(axis=0)
print('Min over height axis:')
print(a)
b = im2s.min(axis=1)
print('Min over width axis:')
print(b)
c = im2s.min(axis=2)
print('Min over color axis:')
print(c)
```

```python
# .data is still just an iterator
print(im2.data[0,0,0], im2.data[0,0,1], im2.data[0,0,2])
print(im2_rgb.data[0,0,0], im2_rgb.data[0,0,1], im2_rgb.data[0,0,2])
```

```python
# Saturate red channel
im3 = im.copy()
for y in range(h):
    for x in range(w):
        im3[y, x, 2] = 255  # Remember:  BGR format
plt.figure()
plt.imshow(im3[:,:,::-1])  # Need to reverse color order, since `imshow` expects RGB
```
